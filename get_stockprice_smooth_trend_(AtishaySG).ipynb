```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import style
```

```python
df_google = pd.read_csv("/content/GOOG.csv")
df_google.head()
```

```python
df_google.info() # check the data types in google csv dataframe
```

```python
df_google.dtypes # short - check the data types in google csv dataframe
```

```python
df_google['date'] = pd.to_datetime(df_google['date'], format='%Y-%m-%d')
df_google.info()
```

```python
style.use("ggplot")
df_google.plot(x="date", y="close", figsize=(16,8))
plt.show()
```

```python
import matplotlib.pyplot as plt
close_avg = df_google["close"].values
window_size = 300
window = np.ones(window_size)/window_size
# Apply the moving average window
closed = np.convolve(close_avg, window,mode="same")
plt.figure(figsize=(16, 8))
plt.plot(df_google["date"], closed, label="Smoothed")
plt.xlabel("date")
plt.ylabel("close")
plt.legend()
plt.show()
```
